```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Data from the top 1500 ranked players in the ATP over the span of 2009 to 2017
#streeteasy = pd.read_csv("https://raw.githubusercontent.com/sonnynomnom/Codecademy-Machine-Learning-Fundamentals/master/StreetEasy/manhattan.csv")

streeteasy = pd.read_csv("tennis_stats.csv")
df = pd.DataFrame(streeteasy)

print(df.head())
```

```python
print(df.columns)
```

```python
x = df[['FirstServe','FirstServePointsWon','FirstServeReturnPointsWon',
'SecondServePointsWon','SecondServeReturnPointsWon','Aces',
'BreakPointsConverted','BreakPointsFaced','BreakPointsOpportunities',
'BreakPointsSaved','DoubleFaults','ReturnGamesPlayed','ReturnGamesWon',
'ReturnPointsWon','ServiceGamesPlayed','ServiceGamesWon','TotalPointsWon',
'TotalServicePointsWon']]

y = df[['Winnings']]

# Training and Test Set

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state=6)

# Multiple linear regression
mlr = LinearRegression()
model=mlr.fit(x_train, y_train)
# Coefficients  - Impact of a variable * "M"
print(model.coef_)
# Intercept - Impact of a variable + "B"
print(model.intercept_)

y_predict = mlr.predict(x_test)

# 2D Graph Scatterplot
plt.scatter(y_test, y_predict, alpha=0.4)
plt.xlabel("Winning")
plt.ylabel("Number of Buildings")
plt.title("Winning at Tennis")
plt.show()
```

```python
# TSS - Total Sum of Squares
# R² is the percentage variation in y explained by all the x variables together
# R² for our model is 0.72, x variables(column) together explain 72% variation(column) in y

print("Train score:")
print(mlr.score(x_train, y_train))
print("Test score:")
print(mlr.score(x_test, y_test))
```

```python
# Rebuild the Model and Predict Price
x = df[['FirstServe','FirstServePointsWon','FirstServeReturnPointsWon',
'SecondServePointsWon','BreakPointsFaced','BreakPointsOpportunities',
'BreakPointsSaved','ReturnGamesPlayed','ReturnGamesWon','ServiceGamesWon',
'TotalServicePointsWon']]

y = df[['Winnings']]

# Training and Test Set

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state=6)

# Multiple linear regression
mlr = LinearRegression()
model= mlr.fit(x_train, y_train)
# Coefficients  - Impact of a variable * "M"
print(model.coef_)
# Intercept - Impact of a variable + "B"
print(model.intercept_)

# TSS - Total Sum of Squares
# R² is the percentage variation in y explained by all the x variables together
# R² for our model is 0.72, x variables(column) together explain 72% variation(column) in y

print("Train score:")
print(mlr.score(x_train, y_train))
print("Test score:")
print(mlr.score(x_test, y_test))

# Score changed
# Train 84.4% -> 84.0%
# Test 82.9% -> 83.4%
```
